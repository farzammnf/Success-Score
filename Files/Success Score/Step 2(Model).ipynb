```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns


# Step 1: Compute differences
data['difference'] = data['home_success_score'] - data['away_success_score']

# Step 2: Check normality of differences
shapiro_test = stats.shapiro(data['difference'])
print(f"Shapiro-Wilk test p-value: {shapiro_test.pvalue}")

if shapiro_test.pvalue > 0.05:
    print("The differences are approximately normally distributed.")
else:
    print("The differences are not normally distributed. Consider using a Wilcoxon Signed-Rank Test.")

# Step 3: One-tailed paired t-test
t_statistic, p_value = stats.ttest_rel(data['home_success_score'], data['away_success_score'], alternative='greater')

# Output results
print(f"Paired t-test statistic: {t_statistic:.3f}")
print(f"Paired t-test one-tailed p-value: {p_value:.5f}")

# Step 4: Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Home success scores are significantly greater than away success scores.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that home success scores are greater.")
    
# Step 5: Plot histograms and box plot
plt.figure(figsize=(12, 6))

# Histogram of Home and Away Scores
plt.subplot(1, 2, 1)
sns.histplot(data['home_success_score'], kde=True, color='blue', label='Home Success Score', alpha=0.6)
sns.histplot(data['away_success_score'], kde=True, color='red', label='Away Success Score', alpha=0.6)
plt.title('Home vs Away Success Scores')
plt.xlabel('Success Score')
plt.legend()

# Boxplot of Home and Away Scores
plt.subplot(1, 2, 2)
sns.boxplot(data=[data['home_success_score'], data['away_success_score']], palette=['blue', 'red'])
plt.xticks([0, 1], ['Home Success Score', 'Away Success Score'])
plt.title('Comparison of Home and Away Success Scores')
plt.ylabel('Success Score')

plt.tight_layout()
plt.show()
```

```python
import joblib
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from sklearn.utils import resample
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# Assuming 'data' is a DataFrame that contains all the feature columns
features = [
    'home_overall', 'home_attack', 'home_midfield', 'home_defence', 'home_international_prestige',
    'home_domestic_prestige', 'home_possession_play_rolling_avg', 'home_counter_attack_rolling_avg',
    'home_high_press_rolling_avg', 'home_direct_play_rolling_avg', 'home_wing_play_rolling_avg',
    'home_set_piece_focus_rolling_avg', 'home_creative_playmaking_rolling_avg', 'home_playing_out_back_rolling_avg',
    'home_low_block_rolling_avg', 'home_high_defensive_line_rolling_avg', 'home_success_score_rolling_avg',
    'home_conceded_goals_rolling_avg', 'home_score_rolling_avg', 'home_points_rolling_avg',
    'home_possession_play_rolling_avg2', 'home_counter_attack_rolling_avg2',
    'home_high_press_rolling_avg2', 'home_direct_play_rolling_avg2', 'home_wing_play_rolling_avg2',
    'home_set_piece_focus_rolling_avg2', 'home_creative_playmaking_rolling_avg2', 'home_playing_out_back_rolling_avg2',
    'home_low_block_rolling_avg2', 'home_high_defensive_line_rolling_avg2', 'home_success_score_rolling_avg2',
    'home_conceded_goals_rolling_avg2', 'home_score_rolling_avg2', 'home_points_rolling_avg2',
    'away_overall', 'away_attack', 'away_midfield', 'away_defence', 'away_international_prestige',
    'away_domestic_prestige', 'away_possession_play_rolling_avg', 'away_counter_attack_rolling_avg',
    'away_high_press_rolling_avg', 'away_direct_play_rolling_avg', 'away_wing_play_rolling_avg',
    'away_set_piece_focus_rolling_avg', 'away_creative_playmaking_rolling_avg', 'away_playing_out_back_rolling_avg',
    'away_low_block_rolling_avg', 'away_high_defensive_line_rolling_avg', 'away_success_score_rolling_avg',
    'away_conceded_goals_rolling_avg', 'away_score_rolling_avg', 'away_points_rolling_avg',
    'away_possession_play_rolling_avg2', 'away_counter_attack_rolling_avg2',
    'away_high_press_rolling_avg2', 'away_direct_play_rolling_avg2', 'away_wing_play_rolling_avg2',
    'away_set_piece_focus_rolling_avg2', 'away_creative_playmaking_rolling_avg2', 'away_playing_out_back_rolling_avg2',
    'away_low_block_rolling_avg2', 'away_high_defensive_line_rolling_avg2', 'away_success_score_rolling_avg2',
    'away_conceded_goals_rolling_avg2', 'away_score_rolling_avg2', 'away_points_rolling_avg2',
]

# Extract the feature data from the DataFrame
feature_data = data[features]
target = data["home_success_score"]

# Normalize features using StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(feature_data)
joblib.dump(scaler, 'scaler.pkl')  # Save the scaler

# Combine features and target into a single DataFrame for easier sampling
data_combined = pd.concat([pd.DataFrame(features_scaled), pd.Series(target, name='target')], axis=1)

# Specify the desired dataset size (e.g., 2x the original dataset size)
desired_size = 10 * len(data_combined)

# Perform oversampling
oversampled_data = resample(
    data_combined,
    replace=True,  # Allow duplicates
    n_samples=desired_size,  # New total size of the dataset
    random_state=42  # For reproducibility
)

# Separate features and target after oversampling
features_oversampled = oversampled_data.drop(columns=['target']).values
target_oversampled = oversampled_data['target'].values

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features_oversampled, target_oversampled, test_size=0.2, random_state=42)

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)



# Print the original feature data shape
print("Original feature data shape:", feature_data.shape)
print("Original target data shape:", target.shape)

# Print the scaled features shape
print("Scaled features shape:", features_scaled.shape)

# Print the oversampled data shape
print("Oversampled features shape:", features_oversampled.shape)
print("Oversampled target shape:", target_oversampled.shape)

# After splitting the data
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

# PyTorch tensor shapes
print("X_train_tensor shape:", X_train_tensor.shape)
print("y_train_tensor shape:", y_train_tensor.shape)
print("X_test_tensor shape:", X_test_tensor.shape)
print("y_test_tensor shape:", y_test_tensor.shape)

# Model Definition with Tanh activation
class DeepNeuralNetwork(nn.Module):
    def __init__(self):
        super(DeepNeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(68, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, 1)
        self.tanh = nn.Tanh()
        self.dropout = nn.Dropout(0.2)
    
    def forward(self, x):
        x = self.tanh(self.fc1(x))
        x = self.dropout(x)
        x = self.tanh(self.fc2(x))
        x = self.dropout(x)
        x = self.tanh(self.fc3(x))
        x = self.fc4(x)  # No activation for the final layer
        return x

# Initialize model, optimizer, loss function, and scheduler
model = DeepNeuralNetwork()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
criterion = nn.L1Loss()  # Using MAE
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=10, verbose=True)

# Training Function
def train_model_with_validation(model, optimizer, train_loader, val_loader, num_epochs=40, early_stopping_patience=500):
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    patience_counter = 0

    for epoch in range(num_epochs):
        model.train()
        running_train_loss = 0.0

        # Training loop
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_train_loss += loss.item() * inputs.size(0)

        train_loss = running_train_loss / len(train_loader.dataset)
        train_losses.append(train_loss)

        # Validation loop
        model.eval()
        running_val_loss = 0.0
        with torch.no_grad():
            for inputs, labels in val_loader:
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                running_val_loss += loss.item() * inputs.size(0)

        val_loss = running_val_loss / len(val_loader.dataset)
        val_losses.append(val_loss)

        print(f"Epoch {epoch+1}/{num_epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            torch.save(model.state_dict(), 'best_model_tanh.pth')
        else:
            patience_counter += 1
            if patience_counter >= early_stopping_patience:
                print("Early stopping triggered.")
                break

        scheduler.step(val_loss)

    return train_losses, val_losses

# Split validation data
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.float32).view(-1, 1)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Train model
train_losses, val_losses = train_model_with_validation(model, optimizer, train_loader, val_loader, num_epochs=40)

# Plot Loss
plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss with Tanh')
plt.legend()
plt.show()

# Evaluate the model
def evaluate_model(model, test_loader):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            all_preds.append(outputs.numpy())
            all_labels.append(labels.numpy())

    all_preds = np.concatenate(all_preds, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)

    mae = mean_absolute_error(all_labels, all_preds)
    mse = mean_squared_error(all_labels, all_preds)
    r2 = r2_score(all_labels, all_preds)

    print(f"MAE: {mae:.4f} | MSE: {mse:.4f} | R²: {r2:.4f}")

evaluate_model(model, test_loader)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define the number of folds
num_folds = 5

# Prepare for k-fold cross-validation
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Store results across all folds
fold_train_losses = []
fold_val_losses = []
fold_mae = []
fold_mse = []
fold_r2 = []

# Perform 5-fold cross-validation
for fold, (train_idx, val_idx) in enumerate(kf.split(features_oversampled, target_oversampled)):
    print(f"Training fold {fold + 1}/{num_folds}")

    # Split data into train and validation for this fold
    X_train_fold, X_val_fold = features_oversampled[train_idx], features_oversampled[val_idx]
    y_train_fold, y_val_fold = target_oversampled[train_idx], target_oversampled[val_idx]

    # Convert to PyTorch tensors
    X_train_tensor = torch.tensor(X_train_fold, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train_fold, dtype=torch.float32).view(-1, 1)
    X_val_tensor = torch.tensor(X_val_fold, dtype=torch.float32)
    y_val_tensor = torch.tensor(y_val_fold, dtype=torch.float32).view(-1, 1)

    # Create DataLoader for this fold
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)

    # Initialize the model, optimizer, and criterion
    model = DeepNeuralNetwork()
    optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
    criterion = nn.L1Loss()  # MAE loss function
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=10, verbose=True)

    # Train the model
    train_losses, val_losses = train_model_with_validation(model, optimizer, train_loader, val_loader, num_epochs=40)

    # Store fold losses
    fold_train_losses.append(train_losses)
    fold_val_losses.append(val_losses)

    # Evaluate the model on the validation set of the current fold
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            all_preds.append(outputs.numpy())
            all_labels.append(labels.numpy())

    all_preds = np.concatenate(all_preds, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)

    # Calculate metrics
    mae = mean_absolute_error(all_labels, all_preds)
    mse = mean_squared_error(all_labels, all_preds)
    r2 = r2_score(all_labels, all_preds)

    fold_mae.append(mae)
    fold_mse.append(mse)
    fold_r2.append(r2)

    print(f"Fold {fold + 1} | MAE: {mae:.4f} | MSE: {mse:.4f} | R²: {r2:.4f}")

# After cross-validation, compute the average metrics
avg_mae = np.mean(fold_mae)
avg_mse = np.mean(fold_mse)
avg_r2 = np.mean(fold_r2)

print(f"\nCross-Validation Results: ")
print(f"Average MAE: {avg_mae:.4f}")
print(f"Average MSE: {avg_mse:.4f}")
print(f"Average R²: {avg_r2:.4f}")

# Optionally, you can plot the loss curves for each fold
plt.figure(figsize=(10, 6))
for i in range(num_folds):
    plt.plot(fold_train_losses[i], color='blue', label=f"Train Loss" if i == 0 else "")
    plt.plot(fold_val_losses[i], color='orange', label=f"Validation Loss" if i == 0 else "")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Across Folds')
plt.legend()
plt.show()
```

```python
import joblib
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from sklearn.utils import resample
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# Assuming 'data' is a DataFrame that contains all the feature columns
features = [
    'home_overall', 'home_attack', 'home_midfield', 'home_defence', 'home_international_prestige',
    'home_domestic_prestige', 'home_possession_play_rolling_avg', 'home_counter_attack_rolling_avg',
    'home_high_press_rolling_avg', 'home_direct_play_rolling_avg', 'home_wing_play_rolling_avg',
    'home_set_piece_focus_rolling_avg', 'home_creative_playmaking_rolling_avg', 'home_playing_out_back_rolling_avg',
    'home_low_block_rolling_avg', 'home_high_defensive_line_rolling_avg', 'home_success_score_rolling_avg',
    'home_conceded_goals_rolling_avg', 'home_score_rolling_avg', 'home_points_rolling_avg',
    'home_possession_play_rolling_avg2', 'home_counter_attack_rolling_avg2',
    'home_high_press_rolling_avg2', 'home_direct_play_rolling_avg2', 'home_wing_play_rolling_avg2',
    'home_set_piece_focus_rolling_avg2', 'home_creative_playmaking_rolling_avg2', 'home_playing_out_back_rolling_avg2',
    'home_low_block_rolling_avg2', 'home_high_defensive_line_rolling_avg2', 'home_success_score_rolling_avg2',
    'home_conceded_goals_rolling_avg2', 'home_score_rolling_avg2', 'home_points_rolling_avg2',
    'away_overall', 'away_attack', 'away_midfield', 'away_defence', 'away_international_prestige',
    'away_domestic_prestige', 'away_possession_play_rolling_avg', 'away_counter_attack_rolling_avg',
    'away_high_press_rolling_avg', 'away_direct_play_rolling_avg', 'away_wing_play_rolling_avg',
    'away_set_piece_focus_rolling_avg', 'away_creative_playmaking_rolling_avg', 'away_playing_out_back_rolling_avg',
    'away_low_block_rolling_avg', 'away_high_defensive_line_rolling_avg', 'away_success_score_rolling_avg',
    'away_conceded_goals_rolling_avg', 'away_score_rolling_avg', 'away_points_rolling_avg',
    'away_possession_play_rolling_avg2', 'away_counter_attack_rolling_avg2',
    'away_high_press_rolling_avg2', 'away_direct_play_rolling_avg2', 'away_wing_play_rolling_avg2',
    'away_set_piece_focus_rolling_avg2', 'away_creative_playmaking_rolling_avg2', 'away_playing_out_back_rolling_avg2',
    'away_low_block_rolling_avg2', 'away_high_defensive_line_rolling_avg2', 'away_success_score_rolling_avg2',
    'away_conceded_goals_rolling_avg2', 'away_score_rolling_avg2', 'away_points_rolling_avg2',
]

# Extract the feature data from the DataFrame
feature_data = data[features]
target = data["home_success_score"]

# Normalize features using StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(feature_data)
joblib.dump(scaler, 'scaler.pkl')  # Save the scaler

# Combine features and target into a single DataFrame for easier sampling
data_combined = pd.concat([pd.DataFrame(features_scaled), pd.Series(target, name='target')], axis=1)

# Specify the desired dataset size (e.g., 2x the original dataset size)
desired_size = 10 * len(data_combined)

# Perform oversampling
oversampled_data = resample(
    data_combined,
    replace=True,  # Allow duplicates
    n_samples=desired_size,  # New total size of the dataset
    random_state=42  # For reproducibility
)

# Separate features and target after oversampling
features_oversampled = oversampled_data.drop(columns=['target']).values
target_oversampled = oversampled_data['target'].values

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features_oversampled, target_oversampled, test_size=0.2, random_state=42)

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Model Definition with Tanh activation
class DeepNeuralNetwork(nn.Module):
    def __init__(self):
        super(DeepNeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(68, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, 1)
        self.tanh = nn.Tanh()
        self.dropout = nn.Dropout(0.3)
    
    def forward(self, x):
        x = self.tanh(self.fc1(x))
        x = self.dropout(x)
        x = self.tanh(self.fc2(x))
        x = self.dropout(x)
        x = self.tanh(self.fc3(x))
        x = self.fc4(x)  # No activation for the final layer
        return x

# Initialize model, optimizer, loss function, and scheduler
model = DeepNeuralNetwork()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
criterion = nn.MSELoss()  # Using MSE
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=10, verbose=True)

# Training Function
def train_model_with_validation(model, optimizer, train_loader, val_loader, num_epochs=100, early_stopping_patience=50):
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    patience_counter = 0

    for epoch in range(num_epochs):
        model.train()
        running_train_loss = 0.0

        # Training loop
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_train_loss += loss.item() * inputs.size(0)

        train_loss = running_train_loss / len(train_loader.dataset)
        train_losses.append(train_loss)

        # Validation loop
        model.eval()
        running_val_loss = 0.0
        with torch.no_grad():
            for inputs, labels in val_loader:
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                running_val_loss += loss.item() * inputs.size(0)

        val_loss = running_val_loss / len(val_loader.dataset)
        val_losses.append(val_loss)

        print(f"Epoch {epoch+1}/{num_epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            torch.save(model.state_dict(), 'best_model_tanh2.pth')
        else:
            patience_counter += 1
            if patience_counter >= early_stopping_patience:
                print("Early stopping triggered.")
                break

        scheduler.step(val_loss)

    return train_losses, val_losses

# Split validation data
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.float32).view(-1, 1)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Train model
train_losses, val_losses = train_model_with_validation(model, optimizer, train_loader, val_loader, num_epochs=100)

# Plot Loss
plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss with Tanh')
plt.legend()
plt.show()

# Evaluate the model
def evaluate_model(model, test_loader):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            all_preds.append(outputs.numpy())
            all_labels.append(labels.numpy())

    all_preds = np.concatenate(all_preds, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)

    mae = mean_absolute_error(all_labels, all_preds)
    mse = mean_squared_error(all_labels, all_preds)
    r2 = r2_score(all_labels, all_preds)

    print(f"MAE: {mae:.4f} | MSE: {mse:.4f} | R²: {r2:.4f}")

evaluate_model(model, test_loader)
```

```python
def evaluate_model_with_boundaries_and_total(model, test_loader):
    model.eval()
    all_preds, all_labels = [], []
    total_loss = 0.0
    
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            total_loss += loss.item() * inputs.size(0)
            all_preds.append(outputs.numpy())
            all_labels.append(labels.numpy())
        
        avg_loss = total_loss / len(test_loader.dataset)
        print(f'Test Loss: {avg_loss:.4f}')
        
        all_preds = np.concatenate(all_preds, axis=0)
        all_labels = np.concatenate(all_labels, axis=0)
        
        r2 = r2_score(all_labels, all_preds)
        mse = mean_squared_error(all_labels, all_preds)
        print(f'R² Score: {r2:.4f}')
        print(f'Mean Squared Error: {mse:.4f}')
        
        # Check correct predictions within the specified boundaries
        boundaries = [(0, 3.8), (3.8, 5.9), (5.9, 10)]
        correct_predictions = {boundary: 0 for boundary in boundaries}
        total_counts = {boundary: 0 for boundary in boundaries}
        total_correct = 0
        total_count = 0

        for actual, pred in zip(all_labels, all_preds):
            for boundary in boundaries:
                lower, upper = boundary
                if lower <= actual < upper:
                    total_counts[boundary] += 1
                    total_count += 1
                    if lower <= pred < upper:
                        correct_predictions[boundary] += 1
                        total_correct += 1  # Increment total correct predictions
                    break
        
        # Calculate and print the percentages for each boundary
        for boundary in boundaries:
            total = total_counts[boundary]
            correct = correct_predictions[boundary]
            percentage = (correct / total) * 100 if total > 0 else 0
            print(f'Boundary {boundary}: {correct}/{total} correct predictions ({percentage:.2f}%)')

        # Calculate and print the total percentage
        total_percentage = (total_correct / total_count) * 100 if total_count > 0 else 0
        print(f'Total correct predictions: {total_correct}/{total_count} ({total_percentage:.2f}%)')

# Evaluate the model with boundary checking and total percentage
evaluate_model_with_boundaries_and_total(model, test_loader)
```

# Feature Importance

```python
from sklearn.metrics import mean_absolute_error
import numpy as np

def permutation_importance(model, test_loader, baseline_score, n_repeats=5):
    """
    Calculates permutation feature importance for the model.

    Parameters:
    - model: Trained PyTorch model.
    - test_loader: DataLoader with test data.
    - baseline_score: Baseline MAE score with all features.
    - n_repeats: Number of times to repeat shuffling per feature.

    Returns:
    - importance_df: DataFrame with features and their importance scores.
    """
    model.eval()
    feature_importances = np.zeros(X_test.shape[1])

    for i in range(X_test.shape[1]):  # Iterate over each feature
        scores = []
        for _ in range(n_repeats):
            X_test_permuted = X_test.copy()
            np.random.shuffle(X_test_permuted[:, i])  # Shuffle only one feature
            
            # Recalculate predictions with the shuffled feature
            X_test_tensor_permuted = torch.tensor(X_test_permuted, dtype=torch.float32)
            test_dataset_permuted = TensorDataset(X_test_tensor_permuted, y_test_tensor)
            test_loader_permuted = DataLoader(test_dataset_permuted, batch_size=32, shuffle=False)
            
            all_preds = []
            all_labels = []
            with torch.no_grad():
                for inputs, labels in test_loader_permuted:
                    outputs = model(inputs)
                    all_preds.append(outputs.numpy())
                    all_labels.append(labels.numpy())

            all_preds = np.concatenate(all_preds, axis=0)
            all_labels = np.concatenate(all_labels, axis=0)
            permuted_score = mean_absolute_error(all_labels, all_preds)
            scores.append(permuted_score)

        # Calculate mean score change due to feature shuffling
        feature_importances[i] = np.mean(scores) - baseline_score

    # Store results in a DataFrame
    feature_importance_df = pd.DataFrame({
        'Feature': features,
        'Importance': feature_importances
    }).sort_values(by='Importance', ascending=False)

    return feature_importance_df

# Calculate baseline MAE score
all_preds = []
all_labels = []
with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = model(inputs)
        all_preds.append(outputs.numpy())
        all_labels.append(labels.numpy())

all_preds = np.concatenate(all_preds, axis=0)
all_labels = np.concatenate(all_labels, axis=0)
baseline_score = mean_absolute_error(all_labels, all_preds)

# Run permutation importance
importance_df = permutation_importance(model, test_loader, baseline_score)
print(importance_df)
```
