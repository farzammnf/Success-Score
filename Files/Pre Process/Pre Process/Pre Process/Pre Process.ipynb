```python
teams = pd.read_csv("male_teams.csv")
teams.head()
```

```python
columns_to_remove = ['captain', 'short_free_kick', 'long_free_kick', 'left_short_free_kick',
    'right_short_free_kick', 'penalties', 'left_corner', 'right_corner', 'def_style',
    'def_team_width', 'def_team_depth', 'def_defence_pressure', 'def_defence_aggression',
    'def_defence_width', 'def_defence_defender_line', 'off_style', 'off_build_up_play',
    'off_chance_creation', 'off_team_width', 'off_players_in_box', 'off_corners',
    'off_free_kicks', 'build_up_play_speed', 'build_up_play_dribbling', 'build_up_play_passing',
    'build_up_play_positioning', 'chance_creation_passing', 'chance_creation_crossing',
    'chance_creation_shooting', 'chance_creation_positioning', 'transfer_budget_eur', 'coach_id',
    'home_stadium', 'rival_team', 'fifa_update'
]
teams = teams.drop(columns=columns_to_remove)
```

```python
import pandas as pd
from fuzzywuzzy import process

# Assume df_cleaned and teams DataFrames are already loaded

# Step 1: Map 'season' to 'fifa_version'
def season_to_fifa_version(season):
    if season == 2021:
        return 21
    if season == 2122:
        return 22
    if season == 2223:
        return 23
    if season == 2324:
        return 24

df_cleaned['fifa_version'] = df_cleaned['season'].apply(season_to_fifa_version)

print("Mapped seasons to FIFA versions:")
print(df_cleaned[['season', 'fifa_version']].tail())
print("Number of null fifa_version entries:", df_cleaned["fifa_version"].isnull().sum())

# Step 2: Prepare teams DataFrames for merging
columns_to_merge = [
    'league_name', 'league_level', 'nationality_id', 'nationality_name',
    'overall', 'attack', 'midfield', 'defence', 'international_prestige',
    'domestic_prestige', 'club_worth_eur', 'starting_xi_average_age',
    'whole_team_average_age'
]

home_team_info = teams[['team_name', 'fifa_version'] + columns_to_merge].copy()
home_team_info.rename(columns={col: 'home_' + col for col in columns_to_merge}, inplace=True)
home_team_info.rename(columns={'team_name': 'home_team'}, inplace=True)

away_team_info = teams[['team_name', 'fifa_version'] + columns_to_merge].copy()
away_team_info.rename(columns={col: 'away_' + col for col in columns_to_merge}, inplace=True)
away_team_info.rename(columns={'team_name': 'away_team'}, inplace=True)

# Step 3: Get unique team names from the teams DataFrame for fuzzy matching
unique_team_names = set(teams['team_name'])

# Step 4: Define the fuzzy matching function
def fuzzy_match_with_comparison(team_name, choices, threshold=80):
    if isinstance(team_name, str):
        match, score = process.extractOne(team_name, choices)
        if score >= threshold:
            return pd.Series([match, team_name, score])  # Return match, original name, and score
    return pd.Series([team_name, team_name, None])  # Return original name if no match found

# Step 5: Apply fuzzy matching to home and away teams in df_cleaned
df_cleaned[['home_team_matched', 'home_team_original', 'home_team_score']] = df_cleaned['home_team'].apply(
    lambda x: fuzzy_match_with_comparison(x, unique_team_names)
)

df_cleaned[['away_team_matched', 'away_team_original', 'away_team_score']] = df_cleaned['away_team'].apply(
    lambda x: fuzzy_match_with_comparison(x, unique_team_names)
)

# Step 6: Replace original team names with matched names for merging
df_cleaned['home_team'] = df_cleaned['home_team_matched']
df_cleaned['away_team'] = df_cleaned['away_team_matched']

# Step 7: Ensure 'fifa_version' is of the same type in all DataFrames
for df in [df_cleaned, home_team_info, away_team_info]:
    if df['fifa_version'].dtype == 'object':
        df['fifa_version'] = df['fifa_version'].astype(int)

# Step 8: Check for duplicates in the teams DataFrames
home_team_info.drop_duplicates(subset=['home_team', 'fifa_version'], inplace=True)
away_team_info.drop_duplicates(subset=['away_team', 'fifa_version'], inplace=True)

# Step 9: Merge home team information
df_merged = df_cleaned.merge(
    home_team_info,
    on=['home_team', 'fifa_version'],
    how='left'
)

# Step 10: Merge away team information
df_final = df_merged.merge(
    away_team_info,
    on=['away_team', 'fifa_version'],
    how='left'
)

# Display the first few rows to check the added columns
print("Merged DataFrame:")
print(df_final[['home_team', 'home_team_original', 'home_team_score', 'away_team', 'away_team_original', 'away_team_score']].head())
```
