```python
import pandas as pd

# Assuming your loaded DataFrame is named 'df'
# Define the columns to be averaged for home and away teams
home_columns = [col for col in df.columns if 'home_' in col and df[col].dtype in ['int64', 'float64']]
away_columns = [col.replace('home_', 'away_') for col in home_columns]

# Initialize a dictionary to store average values for each team
team_averages = {}

# Calculate average values for each team
for team in pd.concat([df['home_team'], df['away_team']]).unique():
    # Selecting home data for the team
    home_data = df[df['home_team'] == team][home_columns]
    
    # Selecting away data for the team and renaming to match home columns
    away_data = df[df['away_team'] == team][away_columns].rename(columns=lambda x: x.replace('away_', 'home_'))
    
    # Calculate the mean for each column only if there is data
    home_mean = home_data.mean() if not home_data.empty else pd.Series([0] * len(home_columns), index=home_columns)
    away_mean = away_data.mean() if not away_data.empty else pd.Series([0] * len(home_columns), index=home_columns)
    
    # Store the mean of both home and away data for each team
    team_averages[team] = ((home_mean + away_mean) / 2).rename(lambda x: x.replace('home_', ''))

# Convert the team_averages dictionary to a DataFrame
team_averages_df = pd.DataFrame(team_averages).T.reset_index()
team_averages_df.columns = ['team'] + [col.replace('home_', '') for col in home_columns]

# Display the resulting DataFrame
team_averages_df
```

```python
# Sort the DataFrame by 'success_score' in descending order
team_averages_df = team_averages_df.sort_values(by='success_score', ascending=False).reset_index(drop=True)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Sample data for demonstration
# extracted_df = pd.DataFrame({'home_team': [...], 'away_team': [...]})
# teams_rating_df = pd.DataFrame({'team': [...], 'overall': [...], 'attack': [...], ...})

# Define a function to get the best match using TF-IDF and cosine similarity
def tfidf_match(team_name, choices):
    # Combine the team name with the choices
    all_teams = [team_name] + choices.tolist()
    
    # Convert team names into TF-IDF vectors
    vectorizer = TfidfVectorizer().fit(all_teams)
    vectors = vectorizer.transform(all_teams)
    
    # Compute cosine similarity between the first vector (team_name) and the rest
    cosine_similarities = cosine_similarity(vectors[0:1], vectors[1:]).flatten()
    
    # Get the index of the best match
    best_match_idx = np.argmax(cosine_similarities)
    
    # Return the best matching team name from choices
    return choices.iloc[best_match_idx]

# Apply the TF-IDF matching function to both the home_team and away_team columns
extracted_df['home_team_matched'] = extracted_df['home_team'].apply(lambda x: tfidf_match(x, teams_rating_df['team']))
extracted_df['away_team_matched'] = extracted_df['away_team'].apply(lambda x: tfidf_match(x, teams_rating_df['team']))

# Merge the teams_rating data for the home_team with suffix '_home'
merged_df = pd.merge(extracted_df, teams_rating_df, left_on='home_team_matched', right_on='team', suffixes=('', '_home'))
# Rename columns to clearly show they are for the home team
merged_df = merged_df.rename(columns={
    'overall': 'home_overall', 'attack': 'home_attack', 'midfield': 'home_midfield', 
    'defence': 'home_defence', 'club_worth': 'home_club_worth', 
    'defence_domestic_prestige': 'home_defence_domestic_prestige', 
    'international_prestige': 'home_international_prestige'
})

# Merge the teams_rating data for the away_team with suffix '_away'
merged_df = pd.merge(merged_df, teams_rating_df, left_on='away_team_matched', right_on='team', suffixes=('', '_away'))
# Rename columns to clearly show they are for the away team
merged_df = merged_df.rename(columns={
    'overall': 'away_overall', 'attack': 'away_attack', 'midfield': 'away_midfield', 
    'defence': 'away_defence', 'club_worth': 'away_club_worth', 
    'defence_domestic_prestige': 'away_defence_domestic_prestige', 
    'international_prestige': 'away_international_prestige'
})

# Now, you can drop the extra 'team' columns used for merging
merged_df = merged_df.drop(columns=['team', 'team_home', 'team_away'], errors='ignore')

# Display the final result
print(merged_df)
```

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import joblib

# Define the Deep Neural Network
class DeepNeuralNetwork(nn.Module):
    def __init__(self):
        super(DeepNeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(32, 2048)
        self.fc2 = nn.Linear(2048, 1024)
        self.fc3 = nn.Linear(1024, 512)
        self.fc4 = nn.Linear(512, 256)
        self.fc5 = nn.Linear(256, 128)
        self.fc6 = nn.Linear(128, 64)
        self.fc7 = nn.Linear(64, 32)
        self.fc8 = nn.Linear(32, 1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        x = self.relu(self.fc4(x))
        x = self.relu(self.fc5(x))
        x = self.relu(self.fc6(x))
        x = self.relu(self.fc7(x))
        x = self.fc8(x)
        x = torch.clamp(x, min=0, max=10)  # Ensure output is between 0 and 10
        return x
    
# Load the trained model and scaler
model = DeepNeuralNetwork()
model.load_state_dict(torch.load('/Users/farzammanafzadeh/Desktop/Tactics/First Phase/deep_neural_network.pth'))
model.eval()

scaler = joblib.load('/Users/farzammanafzadeh/Desktop/Tactics/First Phase/scaler.pkl')

# Assuming `final_df_merged` is loaded as the DataFrame with relevant data columns
feature_columns = [
    'home_overall', 'home_attack', 'home_midfield', 'home_defence', 'home_international_prestige', 
    'home_domestic_prestige', 'home_possession_play', 'home_counter_attack', 'home_high_press', 
    'home_direct_play', 'home_wing_play', 'home_set_piece_focus', 'home_creative_playmaking', 
    'home_playing_out_back', 'home_low_block', 'home_high_defensive_line', 'away_overall', 'away_attack', 
    'away_midfield', 'away_defence', 'away_international_prestige', 'away_domestic_prestige', 
    'away_possession_play', 'away_counter_attack', 'away_high_press', 'away_direct_play', 'away_wing_play', 
    'away_set_piece_focus', 'away_creative_playmaking', 'away_playing_out_back', 'away_low_block', 
    'away_high_defensive_line'
]

# Prepare the new DataFrame to store results
predicted_results = final_df_merged[['league', 'season', 'game', 'round', 'week', 'date', 
                                     'home_team', 'away_team']].copy()

# Placeholder list for predictions
home_success_scores = []

# Iterating over each row in the final_df_merged to make predictions
for _, row in final_df_merged.iterrows():
    # Extracting the features for the model prediction
    features = row[feature_columns].values.reshape(1, -1)
    
    # Normalizing the features
    features_scaled = scaler.transform(features)
    
    # Converting the features to a PyTorch tensor
    features_tensor = torch.tensor(features_scaled, dtype=torch.float32)
    
    # Predicting the home success score
    with torch.no_grad():
        home_success_score = model(features_tensor).item()
        home_success_score_rounded = round(home_success_score, 2)
        home_success_scores.append(home_success_score_rounded)

# Add predicted home success scores and calculated away success scores to the DataFrame
predicted_results['home_success_score'] = home_success_scores
predicted_results['away_success_score'] = [round(10 - score, 2) for score in home_success_scores]

# Display the results
print(predicted_results.head())
```

```python
import pandas as pd

# Assuming final_df_merged is your main DataFrame
# Define functions for each outcome (win, draw, loss)
def calculate_win(success_score):
    return 1 if success_score >= 5.9 else 0

def calculate_draw(success_score):
    return 1 if 3.9 < success_score < 5.9 else 0

def calculate_loss(success_score):
    return 1 if success_score <= 3.9 else 0

# Calculate games played, wins, draws, and losses for home and away teams
final_df_merged['home_games'] = 1
final_df_merged['away_games'] = 1

final_df_merged['home_wins'] = final_df_merged['home_success_score'].apply(calculate_win)
final_df_merged['away_wins'] = final_df_merged['away_success_score'].apply(calculate_win)

final_df_merged['home_draws'] = final_df_merged['home_success_score'].apply(calculate_draw)
final_df_merged['away_draws'] = final_df_merged['away_success_score'].apply(calculate_draw)

final_df_merged['home_losses'] = final_df_merged['home_success_score'].apply(calculate_loss)
final_df_merged['away_losses'] = final_df_merged['away_success_score'].apply(calculate_loss)

# Create a DataFrame for home and another for away statistics
home_stats = final_df_merged[['home_team', 'home_points', 'home_games', 'home_wins', 'home_draws', 'home_losses']].rename(
    columns={'home_team': 'team', 'home_points': 'points', 'home_games': 'games', 'home_wins': 'wins', 'home_draws': 'draws', 'home_losses': 'losses'})
away_stats = final_df_merged[['away_team', 'away_points', 'away_games', 'away_wins', 'away_draws', 'away_losses']].rename(
    columns={'away_team': 'team', 'away_points': 'points', 'away_games': 'games', 'away_wins': 'wins', 'away_draws': 'draws', 'away_losses': 'losses'})

# Concatenate home and away statistics
all_stats = pd.concat([home_stats, away_stats], ignore_index=True)

# Group by team and sum the statistics
ranking_table = all_stats.groupby('team').sum().sort_values(by='points', ascending=False).reset_index()

# Display the ranking table with games, points, wins, draws, and losses
print(ranking_table)
```
